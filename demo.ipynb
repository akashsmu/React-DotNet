```python
import json
import pandas as pd
```

```python
data = open('ontology.json')
data =json.load(data)
data
```

```python
ids = []
for item in data:
    ids.append(item["id"])

humans = []
humanIndex = ids.index("/t/dd00013")
for i in range(humanIndex+1):
        humans.append(data[i]["name"])

animals = []
animalsStartIndex = ids.index("/m/0jbk")
animalsEndIndex = ids.index("/m/032n05")
for i in range(animalsStartIndex,animalsEndIndex+1):
      animals.append(data[i]["name"])


music = []
musicStartIndex = ids.index("/m/04rlf")
musicEndIndex = ids.index("/t/dd00037")
for i in range(musicStartIndex, musicEndIndex+1):
      music.append(data[i]["name"])

naturalSound = []
naturalSoundStartIndex = ids.index("/m/059j3w")
naturalSoundEndIndex = ids.index("/m/0fjy1")
for i in range(naturalSoundStartIndex, naturalSoundEndIndex+1):
      naturalSound.append(data[i]["name"])

soundOfThings = []
soundOfThingsStartIndex = ids.index("/t/dd00041")
soundOfThingsEndIndex = ids.index("/m/0dv3j")
for i in range(soundOfThingsStartIndex,soundOfThingsEndIndex+1):
      soundOfThings.append(data[i]["name"])


ambiguous = []
ambiguousStartIndex = ids.index("/t/dd00098")
ambiguousEndIndex = ids.index("/m/05szwhw")
for i in range(ambiguousStartIndex,ambiguousEndIndex+1):
      ambiguous.append(data[i]["name"])


envBackground = []
envBackgroundStartIndex = ids.index("/t/dd00123")
envBackgroundEndIndex = ids.index("/m/04zc0")
for i in range(envBackgroundStartIndex,envBackgroundEndIndex+1):
      envBackground.append(data[i]["name"])


```

```python
choices = {}
choices[animals[0]] = animals[1:]
choices[music[0]] = music[1:]
choices[humans[0]] = humans[1:]
choices[naturalSound[0]] = naturalSound[1:]
choices[soundOfThings[0]] = soundOfThings[1:]
choices[ambiguous[0]] = ambiguous[1:]
choices[envBackground[0]] = envBackground[1:]
```

```python
keys = list(choices.keys())
keys
```

```python
with open("choices.json", "w") as f:
    json.dump(choices,f) 
```
